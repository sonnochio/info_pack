### Dual Token & Burning System
<!--  -->
<img src="img/token_eco.png" style="float: left ;margin-right: 10px;margin-top: 25px;"   width="700"/> 





**$SQUAD Governence Token**: 
- **Synthesising/Upgrading NFTs** : of which **30% is burned**Ôºå remaining **50% to DAO pool**

**$ECG Game Token**: 

- **Synthesing/Upgrading** NFTs: Results in mass consumption of \\$ECG , **burn 100% \\$ECG** consumed.

- **Mystery Chest** : Results in mass consumption of \\$ECG, **burn 100% \\$ECG** consumed.

- **Renting** NFTs: **burn 50% $ECG** consumed, remaining **50% to DAO pool**.

- **Transaction Fee**: We charge a **1\% transaction fee** for each $ECG withdrawal. **50% is burned**, remaining **50% allocates to DAO pool**.  

#### Addition Info on Tokenomics 

```python
token=pd.read_csv('token_distribution.csv')
```

```python
pd.set_option('display.max_columns',None)
# pd.set_option('display.max_rows', None)
# pd.set_option('display.width',None)
pd.set_option('display.max_colwidth', None)
token_purp=pd.read_csv("token_purp.csv").set_index("Allocation")
token_purp
```

```python
token_distribution_value=token.iloc[:,10:58].applymap(lambda x: x.replace('%', '')).astype('float').cumsum(axis=1)
token_distribution_value['Allocation']=token_purp.index
token_distribution_value=token_distribution_value.set_index('Allocation').rename(columns={'Month1':'Month 1'})
token_distribution_value
```

```python
(token['Percentage'].apply(lambda x: x.replace('%', '')).astype('float'))
```

```python
value_perc=token_distribution_value.round(2).astype('string').applymap(lambda x: x+'%')
value_perc
```

```python
value_perc.to_csv('cumsum.csv')
```

```python
#building the data frame
df=pd.DataFrame()
df['Investor Tokens']=(token_distribution_value.iloc[[8,9],:36]).sum()
df['Team Tokens']=token_distribution_value.iloc[10,:36]
df['Community Tokens']=token_distribution_value.iloc[[0,3,4,5,7,11],:36].sum()
df['Total Release']=token_distribution_value.iloc[:,:36].sum()
column_dict={}
for i in range(48):
    column_dict[f'Month {i+1}']=f'{i+1}'
    

df_new=df.rename(column_dict).rename({'Month1':'1'})
df_new
```

```python
token_distribution_value=token.iloc[:,10:58].applymap(lambda x: float(x.replace('%',''))/100)
token_distribution_value
```
